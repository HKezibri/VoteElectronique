```python
def Password():#-----------------fonction pour générer aléatoirement le mote de pass pendant l'inscription
    import random as r
    #--------------------------------------pour l’alphabet en majuscule
    PX = ''
    for i in range(4):
        Xr = r.randint(65, 90)
        PX += chr(Xr)
    #-------------------------------------pour les nombres entre 0 et 9 
    DX = ''
    for e in range(4):
        Dr = r.randint(0, 9)
        DX += str(Dr)
    return (PX+'@'+DX)
```

```python
def Login(n):#----------------------------------fonction pour donner un Login lors de l'inscription
    
    e = 'E000'
    if n <= 9:
        e = e[:-1]+str(n)
        return e
    elif n <= 99:
        e = e[:-2]+str(n)
        return e
    elif n <= 999:
        e = e[:-3]+str(n)
        return e
    else:
        return "Non autorisé !!! "
    
```

```python
def Test_inscription(nom, prenom):    #------------------prendre le nom et le prénom qui retourne la fonction test_nom_prenom
    with open("ELECTEURS.txt", "r") as G:    #---------- et faire l'inscription aprés avoir tester que l'électeur
        g = G.readlines()                    #------------ n'est pas deja inscrit
        compteur = 0
        for e in g:
            T = e.split(' ')
            if nom.upper() == T[2] and prenom.upper() == T[3][:-1]:
                compteur+=1
        
        if compteur != 0:
            print('\t Vous êtes déjà inscrit !! \t\t Merci')
            Menu()###
        if compteur == 0 :
            with open("ELECTEURS.txt", "a") as F:
                with open("ELECTEURS.txt", "r") as E:
                    N = E.readlines()
                F.write(Password()+' '+Login(len(N))+' '+nom.upper()+' '+prenom.upper()+'\n')
            with open("ELECTEURS.txt", "r") as G:
                g = G.readlines()                    
                print("-------------Inscription réussie -------------")
                print("Voici vous informations :"+"\n"+"Login : "+g[len(g)-1].split(' ')[1]+"\t"+
                      "Password : "+g[len(g)-1].split(' ')[0])   
                Menu()###                
```

```python
def Test_Nom_Prenom() :    #----------------donner un vrai nom et prénom 
    while True :
        nom = input("Nom  :  ")
        prenom = input("Prénom  :  ") 
        List_Nom = [e for e in nom.upper()]
        List_Prenom = [i for i in prenom.upper()]
        L=[chr(j) for j in range(65,91)]
        cpt = 0
        if nom == '' or prenom == '' or len(nom) < 3 or len(prenom) < 3 or len(nom) > 15 or len(prenom) > 15 :
            print("\t\t--- La taille de Nom et/ou Prénom est inaccéptable !! --- ")
        if 3 < len(nom) < 15 and 3 < len(prenom) < 15 :
            for e in List_Nom :
                if e in L:
                    cpt+=1
            for e in List_Prenom :
                if e in L:
                    cpt+=1 
            if cpt == len(List_Prenom+List_Nom):
                return nom,prenom
                break
        else:
            print('Réessayez..')
```

```python
def Inscription():
    _ = Test_Nom_Prenom()
    Test_inscription(_[0], _[1])
   
```

```python
def Candidats():    #----------------------  donne la liste des candidats touus simplement
    with open("CANDIDATS.txt", "r") as C:
        L = C.readlines()
        print('___________________________________\n')
        for e in L[:-1]:
            print('\t'+e.split('\n')[0].split(' ')[0]+" --> "+e.split('\n')[0].split(' ')[1]+" "+e.split('\n')[0].split(' ')[2])
        print('\t___________________________')
        print('___________________________________\n')
        
```

```python
def Test_Choix_Existe(choix):    #---------------- test le choix de l'electeur est qu'il existe ou non
    L_choix = ['C1', 'C2', 'C3', 'C4']        #---- apres retourne le choix 
    if choix.upper() in L_choix :
        return choix.upper()
    else:
        while choix.upper() not in L_choix :
            print("Inacceptable !! << Vous devriez revoir la liste des candidats. >>")
            newChoix = input("Vous votez pour qui? ")
            choix = newChoix
        return choix.upper()
```

```python
def Pas_confirm():    #------------------------------- stocke les votant apres de vrai confirmation !!!
    import pickle
    choix = input("Vous votez pour qui? ")
    Test_Choix_Existe(choix)
    confirmChoix = input("Confirmer votre choix [YES/NO] --> (Y / N) : ")
    L_confirmChoix = ['YES', 'NO', 'Y', 'N']
    while True : 
        if confirmChoix.upper() not in L_confirmChoix :
            print("Inacceptable !!!")
            confirmChoix = input("Confirmer votre choix [YES/NO] --> (Y / N) : ")
        else:
            if confirmChoix.upper() == 'N' or confirmChoix.upper() == 'NO' :
                choix = input("Quel est votre choix : ")
                Test_Choix_Existe(choix)
                confirmChoix = input("Confirmer votre choix [YES/NO] --> (Y / N) : ")
            if confirmChoix.upper()  == 'Y' or confirmChoix.upper() == 'YES' :
                with open("RESULTAT.txt", "rb") as R_Vote :
                    stock_R = pickle.Unpickler(R_Vote)
                    Dict_V = stock_R.load()
                CH = Test_Choix_Existe(choix)
                if CH == 'C1' :
                    Dict_V['C1'] += 1
                if CH == 'C2' :
                    Dict_V['C2'] += 1
                if CH == 'C3' :
                    Dict_V['C3'] += 1
                if CH == 'C4' :
                    Dict_V['C4'] += 1
                with open("RESULTAT.txt", "wb") as W_Vote :
                    stock_R = pickle.Pickler(W_Vote)
                    stock_R.dump(Dict_V)
                print('\nEffectué avec succés \n\n\t\t\t\t\t Merci !')
                Menu()###
                break
```

```python
def code_Pwd_correct(code, Password):
    import pickle
    with open("ELECTEURS.txt", "r") as G:
        g = G.readlines()
        c = 0
        for e in g:
            T = e.split(' ')
            if ((code.upper() == T[1])  and (Password.upper() == T[0])) :
                for i in range(len(g)):
                    if code.upper() in g[i] and Password.upper() in g[i]:   #--- code & pwd existe
                        t = g[i].split(' ')
                        print("\nBienvenu "+t[2]+" "+t[3][:-1]+"\nListe des candidats : ")                
                        Candidats()
                        Pas_confirm()  
                        break
                        #Menu()
                break
            if code.upper() != T[1]  or Password.upper() != T[0] :
                c+=1
            if c == len(g) :
                print("Code et/ou password incorrect !!! \n"+"Vous êtes sûr d'être inscrit à la liste éléctorale !")
                Menu()###
```

```python
def first_vote(code):
    with open("PREMIER_VOTE.txt", "a") as PV:
        PV.write(code.upper()+',')
```

```python
def Voter():
        import pickle
        code = input("Votre Code/Login : ")
        Password = input("Votre Password : ")
        with open("PREMIER_VOTE.txt", "r") as PV:
            All_Logine = PV.readline()
            if code.upper() in All_Logine :
                print("Vous avez deja voté ou vos informations ne sont pas correcte !!\t\t\t\t Merci")
                Menu()###
            else:
                code_Pwd_correct(code, Password)
                first_vote(code)
```

```python
def Menu():
    MN()
    CX = input('Tappez Votre ChoiX : \t')
    x = [chr(e) for e in range(49, 54)]
    while CX not in x :
        print(' \t\t\t\t\t!! Consulter Le MENU !!')
        CX = input('Tappez Votre ChoiX : \t')
    n = eval(CX)
    while True :
        print('\n**********************************************************************************************************')
        if n < 1 or n > 5 :
            print('\t\t\t\t!!! Veuillez consulter le MENU  !!! ')
            n = eval(input('Tappez Votre ChoiX : \t'))
        if n == 1 :
            Inscription()
        if n == 2 :
            Candidats()
            Menu()
        if n == 3:
            Voter()
        if n == 4:
            Statistiques()
            Menu()
        if n == 5 :
            print('\t\t\t\t\t\t\t\t\t\t\t\t\t   MERCI :)')
            break
        break
```
